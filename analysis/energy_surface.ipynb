```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
Delta_t = 0.00580144
lambda_ = 1e-3
tau = 1


# ==== Energy function ====


def energy(gamma):
    term1 = np.dot(gamma, phi)
    term2 = 0.5 * Delta_t * gamma.T@D.T@M@D@gamma
    term3 = np.sum((Delta_t / lambda_) * (ell / tau)
                   * np.exp(-lambda_ * (L @ gamma)))
    return term1 + term2 + term3


def gradient(gamma):
    term1 = phi
    term2 = Delta_t * D.T @ M @ D @ gamma
    exp_term = np.exp(-lambda_ * (L @ gamma))
    term3 = -Delta_t * L.T@((ell / tau) * exp_term)
    return term1 + term2 + term3


def res(gamma, phi):
    proj = np.where(gamma > 0, phi, np.minimum(phi, 0))
    return np.linalg.norm(proj, ord=np.inf)


def bbpgd(gamma0):

    gamma_path = [gamma0]
    gradient_history = [gradient(gamma0)]
    res_history = [res(gamma_path[-1], gradient_history[-1])]
    alpha_history = [1/res_history[-1]]
    while True:
        gamma_new = np.maximum(
            # project to non-negative
            gamma_path[-1] - alpha_history[-1] * gradient(gamma_path[-1]), 0)
        gamma_path.append(gamma_new)

        grad_new = gradient(gamma_new)
        gradient_history.append(grad_new)

        res_new = res(gamma_new, grad_new)
        res_history.append(res_new)

        if res_new < 1e-3:
            break

        delta_gamma = gamma_new - gamma_path[-2]
        delta_grad = grad_new - gradient_history[-2]

        alpha_new = np.dot(delta_gamma, delta_grad) / \
            np.dot(delta_grad, delta_grad)
        alpha_history.append(alpha_new)

    return np.array(gamma_path)


# ==== Evaluate grid ====
g1 = np.linspace(-1, 4, 100)
g2 = np.linspace(-1, 4, 100)
G1, G2 = np.meshgrid(g1, g2)

E = np.zeros_like(G1)
min_valid_energy = np.inf
min_valid_gamma = None
for i in range(G1.shape[0]):
    for j in range(G1.shape[1]):
        E[i, j] = energy(np.array([G1[i, j], G2[i, j]]))
        if E[i, j] < min_valid_energy and G1[i, j] >= 0 and G2[i, j] >= 0:
            min_valid_energy = E[i, j]
            min_valid_gamma = (G1[i, j], G2[i, j])

# ==== Plot ====
fig = plt.figure(figsize=(10, 5))

ax1 = fig.add_subplot(1, 2, 1, projection='3d')
ax1.plot_surface(G1, G2, E, cmap='viridis', linewidth=0, antialiased=True)
ax1.set_xlabel(r'$\gamma_1$')
ax1.set_ylabel(r'$\gamma_2$')
ax1.set_zlabel(r'$E(\gamma)$')
ax1.set_title('Energy Surface')


ax2 = fig.add_subplot(1, 2, 2)
levels = np.linspace(0, 10, 101)
cont = ax2.contourf(G1, G2, E, 25, cmap='viridis')
ax2.set_xlabel(r'$\gamma_1$')
ax2.set_ylabel(r'$\gamma_2$')
ax2.set_title('Energy Contours $E(\gamma)$')
fig.colorbar(cont, ax=ax2, label=r'$E(\gamma)$')

# draw in x and y axis, as we cannot do projected gradient descent
ax2.axhline(0, color='white', linestyle='--', linewidth=1)
ax2.axvline(0, color='white', linestyle='--', linewidth=1)

# ==== Gradient descent path ====
gamma_path = bbpgd(np.array([3, 3]))
ax2.plot(gamma_path[:, 0], gamma_path[:, 1], color='red',
         marker='o', markersize=5, label='Gradient Descent Path', linewidth=2)


# annotate with final gamma
final_gamma = gamma_path[-1]


ax2.annotate(f'$\\gamma^* = ({final_gamma[0]:.2f}, {final_gamma[1]:.2f})$',
             xy=(final_gamma[0], final_gamma[1]),
             xytext=(final_gamma[0], final_gamma[1]+2),  # text offset
             ha='center', fontsize=12,
             bbox=dict(boxstyle='round,pad=0.2',
                       facecolor='yellow', alpha=0.9),
             # line pointing to gamma*
             arrowprops=dict(arrowstyle='->', color='yellow', lw=2))


plt.tight_layout()
plt.show()

extent = ax2.get_tightbbox(fig.canvas.get_renderer()).transformed(fig.dpi_scale_trans.inverted())
fig.savefig('energy_contours_ax2.png', dpi=400, bbox_inches=extent)
```
